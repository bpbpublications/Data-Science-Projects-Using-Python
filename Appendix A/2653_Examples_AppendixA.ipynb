```python
#Appendix1.1
import pandas as pd
# Read a csv file
df = pd.read_csv("d:\\students.csv")
# Average score for each student
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
# Find the student who scored highest marks
top_scorer = df.loc[df["Average"].idxmax()]
# Sort by average
sorted_df = df.sort_values(by="Average", ascending=False)
print("Student Data with Averages:\n", df)
print("\nTop Scorer:\n", top_scorer)
print("\nStudents Sorted by Average:\n", sorted_df)
```

```python
#Appendix A1.2
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("d:\\sales.csv")
# Total sales by product using groupby()
product_sales = df.groupby("Product")["Sales"].sum()
# Daily sales
daily_sales = df.groupby("Day")["Sales"].sum()
print("Sales by Product:\n", product_sales)
print("\nSales by Day:\n", daily_sales)
# Visualization
daily_sales.plot(kind="bar", title="Daily Sales")
plt.show()
```

```python
#Appendix A1.3
import pandas as pd
df = pd.read_csv("d:\\employees.csv")
# Average salary by department
avg_salary = df.groupby("Department")["Salary"].mean()
# Highest paid employee
highest_paid = df.loc[df["Salary"].idxmax()]
print("Average Salary by Department:\n", avg_salary)
print("\n Highest Paid Employee:\n", highest_paid)

```

```python
#Appendix A1.4
import pandas as pd
# List of sales amounts
sales = [250, 350, 150, 450, 200]
# Convert list into a DataFrame
df = pd.DataFrame(sales, columns=["Sales"])
print("Sales Data:")
print(df)
print("\nTotal Sales:", df["Sales"].sum())
```

```python
#Appendix A1.5
import pandas as pd
# List of tuples (Name, Age, Department)
employees = [
    ("Anil", 25, "HR"),
    ("Binu", 30, "IT"),
    ("Chetan", 28, "Finance")
]
# Convert list of tuples into DataFrame
df = pd.DataFrame(employees, columns=["Name", "Age", "Department"])
print("Employee Data:")
print(df)
# Find average age of employees
print("\nAverage Age:", df["Age"].mean())
```

```python
#Appendix A1.6
import pandas as pd
# Dictionary with lists
data = {
    "Name": ["Divya", "Sunil", "Vinod"],
    "Marks": [80, 91, 70],
    "Subject": ["Math", "Science", "English"]
}
# Convert dictionary into DataFrame
df = pd.DataFrame(data)
print("Student Data:")
print(df)
# Find maximum marks
print("\nHighest Marks:", df["Marks"].max())
```

```python
#Appendix A1.7
import pandas as pd
# Load CSV file
df = pd.read_csv("d:\\library.csv")
# Show the data
print("Data of Library Books:\n", df)
# Most borrowed book
print("\nMost Borrowed Book:", df.loc[df['Borrowed'].idxmax(), 'Book'])
# Average borrowed count
print("Average Borrowed:", df['Borrowed'].mean())
# Most popular genre
print("Most Popular Genre:")
print(df.groupby("Genre")['Borrowed'].sum().idxmax())
```

```python
#Appendix A1.8
import pandas as pd
# Load CSV file
df = pd.read_csv("d:\\sports.csv")
print("Sports Data:\n", df)
# Highest score player
print("\nTop Player:", df.loc[df['Score'].idxmax(), 'Player'])
# Average score by sport
print("\nAverage Score by Sport:\n", df.groupby("Sport")['Score'].mean())
# Count of players in each sport
print("\nPlayers per Sport:\n", df['Sport'].value_counts())
```

```python
#ExampleA1.9
import pandas as pd
# Load CSV file
df = pd.read_csv("d:\\social.csv")
print("Social Media Data:\n", df)
# Platform with most likes
print("\nMost Liked Platform:", df.loc[df['Likes'].idxmax(), 'Platform'])
# Average shares
print("Average Shares:", df['Shares'].mean())
# Platform with highest total engagement (Likes + Shares)
df['Total'] = df['Likes'] + df['Shares']
print("Most Popular Platform:", df.loc[df['Total'].idxmax(), 'Platform'])
```

```python
#ExampleA1.10
import pandas as pd
#  Load the Iris dataset
df = pd.read_csv("d:\\iris.csv")
print("First 5 Rows of Iris Dataset:\n", df.head())
# Count of each species
print("\nCount of each species:\n", df['species'].value_counts())
#  Average petal length by species
print("\nAverage Petal Length by Species:\n", df.groupby('species')['petal_length'].mean())
# Species with maximum average sepal width
avg_sepal_width = df.groupby('species')['sepal_width'].mean()
print("\nSpecies with Max Average Sepal Width:", avg_sepal_width.idxmax())
# Minimum and Maximum values of each feature
print("\nMinimum values of each column:\n", df.min(numeric_only=True))
print("\nMaximum values of each column:\n", df.max(numeric_only=True))
```
