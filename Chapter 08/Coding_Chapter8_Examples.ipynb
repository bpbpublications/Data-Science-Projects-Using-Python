```python
# Example 8.7

import pandas as pd
# Creating a DataFrame with missing values and duplicates
data = pd.DataFrame({
    'Name': ['Ajay', 'Binu', None, 'Ajay', 'Chetan'],
    'Age': [25, None, 22, 25, 22]
})
# Step 1: Removing rows with missing values
cleaned_data = data.dropna()
print("Data after dropping missing values:")
print(cleaned_data)
# Step 2: Filling missing values in 'Age' with the column's mean
data_filled = data.copy()
data_filled['Age'] = data_filled['Age'].fillna(data_filled['Age'].mean())
print("\nData after filling missing Age values with mean:")
print(data_filled)
# Step 3: Removing duplicate rows
unique_data = data_filled.drop_duplicates()
print("\nData after removing duplicates:")
print(unique_data)
# Step 4: Filtering rows where Age is 25 or more
filtered_data = unique_data[unique_data['Age'] >= 25]
print("\nFiltered data (Age >= 25):")
print(filtered_data)
```

```python
# Example 8.20

import pandas as pd
# Load the CSV file
data = pd.read_csv('D:\\students.csv')
# Display the first 5 rows (optional, for context)
print("First 5 rows of the dataset:")
print(data.head())
# Mean of each subject
print("\nMean scores:")
print(data[['Math', 'Science', 'English']].mean())
# Median of each subject
print("\nMedian scores:")
print(data[['Math', 'Science', 'English']].median())
# Mode of each subject
print("\nMode scores:")
print(data[['Math', 'Science', 'English']].mode())
# Full statistical summary
print("\nDescriptive statistics:")
print(data[['Math', 'Science', 'English']].describe())
```
