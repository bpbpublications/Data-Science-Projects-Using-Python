```python
# Example 9.2

import pandas as pd
# Step 1: Create a dataset with missing values
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math_Marks": [78, None, 92, 70, 88],
    "English_Marks": [80, 75, None, 72, 90]
}
df = pd.DataFrame(data)
print("Original Dataset with Missing Values:")
print(df)
# Step 2: Check for missing values
print("\nMissing Values Count:")
print(df.isnull().sum())
# Step 3: Fill missing values with the mean
df["Math_Marks"].fillna(df["Math_Marks"].mean(), inplace=True)
df["English_Marks"].fillna(df["English_Marks"].mean(), inplace=True)
print("\nDataset After Filling Missing Values with Mean:")
print(df)
```

```python
# Example 9.3

import pandas as pd
# Step 1: Create a dataset with missing values
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math_Marks": [78, None, 92, 70, 88],
    "English_Marks": [80, 75, None, 72, 90]
}
df = pd.DataFrame(data)
print("Original Dataset with Missing Values:")
print(df)
# Step 2: Check for missing values
print("\nMissing Values Count:")
print(df.isnull().sum())
# Step 3: Fill missing values with the median
df["Math_Marks"].fillna(df["Math_Marks"].median(), inplace=True)
df["English_Marks"].fillna(df["English_Marks"].median(), inplace=True)

print("\nDataset After Filling Missing Values with Median:")
print(df)
```

```python
# Example 9.4

import pandas as pd
# Step 1: Create a dataset with missing values
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math_Marks": [78, None, 92, 70, 88],
    "English_Marks": [80, 75, None, 72, 90]
}
df = pd.DataFrame(data)
print("Original Dataset with Missing Values:")
print(df)
# Step 2: Check for missing values
print("\nMissing Values Count:")
print(df.isnull().sum())
# Step 3: Fill missing values using a constant value (e.g., 0 or any default)
df["Math_Marks"].fillna(0, inplace=True)
df["English_Marks"].fillna(0, inplace=True)
print("\nDataset After Filling Missing Values with Constant Value:")
print(df)
```

```python
# Example 9.10
import pandas as pd
# Create a dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Gender": ["Female", "Male", "Male", "Male", "Female"],
    "Grade": ["A", "B", "A", "C", "B"]
}
# Convert to a DataFrame
df = pd.DataFrame(data)
# Display the dataset
print("Original Dataset:")
print(df)
# Select the first 3 rows and the "Gender" column
gender_data = df.iloc[:3, 1]
print("\nFirst 3 rows of the 'Gender' column:")
print(gender_data)
# Select rows 2 to 4 and the last two columns
subset = df.iloc[1:4, 1:]
print("\nRows 2 to 4 and the last two columns:")
print(subset)
```

```python
# Example 9.15
import pandas as pd
# Sample data
data = {'Student': ['A','B','C','D','E'],
        'Marks': [55,68,89,75,85]}
df = pd.DataFrame(data)
# Define bins and labels
bins = [0,40,70,100]
labels = ['Poor','Average','Good']
# Apply binning
df['Grade']= pd.cut(df['Marks'], bins=bins, labels=labels)
print(df)
```

```python
# Example 9.16
import pandas as pd
# Sample data
data = {
    'Name': ['A','B','C','D','E'],
    'Age': [10, 17, 26, 60, 41]
}

df = pd.DataFrame(data)
# Define bins and labels
bins = [0,13,20,59,100]
labels = ['Child','Teenager','Adult','Senior']
# Perform binning
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df)
```
