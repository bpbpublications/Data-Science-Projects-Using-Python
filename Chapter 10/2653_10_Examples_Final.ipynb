```python
# Example 10.1
import pandas as pd
import matplotlib.pyplot as plt
# Creating a simple DataFrame
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 320]
})
# Plotting the data using the built-in plot() method
data.plot(x='Month', y='Sales', kind='line', marker='o', color='blue', title='Monthly Sales')
# Displaying the plot
plt.xlabel('Month')
plt.ylabel('Sales in Units')
plt.grid(True)
plt.show()
```

```python
#Example 10.2
import pandas as pd
import matplotlib.pyplot as plt
# Creating a DataFrame with product sales data
data = pd.DataFrame({
    'Fruits': ['Apples', 'Bananas', 'Cherries', 'Oranges'],
    'Quantity Sold': [120, 90, 60, 30]
})
# Plotting a vertical bar chart
data.plot(x='Fruits', y='Quantity Sold', kind='bar', color='orange', title='Sale of Fruits')
# Adding axis labels
plt.xlabel('Fruits')
plt.ylabel('Units Sold')
# Displaying the plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Example 10.3
import pandas as pd
import matplotlib.pyplot as plt
# Creating a DataFrame for course enrollment
data = pd.DataFrame({
    'Course': ['Mathematics', 'Physics', 'Computer Science', 'English', 'Biology'],
    'Enrolled Students': [45, 30, 60, 25, 40]
})
# Generating a pie chart
plt.figure(figsize=(6,6))
plt.pie(data['Enrolled Students'], labels=data['Course'], autopct='%1.1f%%', startangle=90)
plt.title('Course-wise Student Enrollment')
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```

```python
#Example10.9
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame for monthly enrollment data
df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Enrollments': [40, 50, 80, 60, 100]
})

# Plotting the data
df.plot(x='Month', y='Enrollments', kind='bar', color='skyblue', legend=False)
plt.title('Monthly Enrollments')      # Adding a title
plt.xlabel('Month')                   # Customizing X-axis label
plt.ylabel('Number of Enrollments')   # Customizing Y-axis label
plt.xticks(rotation=0)                # Keeping month names horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)   # Adding grid lines for clarity
plt.legend(['Enrollments'], title='Data Type')  # Custom legend with a title
plt.tight_layout()
plt.show()
```

```python
#Example 10.10
import pandas as pd
import matplotlib.pyplot as plt

# Data of attendance over 5 school days
data = pd.DataFrame({
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Attendance': [45, 47, 44, 48, 46]
})

# Line plot with customized color, style, and marker
plt.plot(data['Day'], data['Attendance'],
         color='green', linestyle='-.', marker='o', label='Attendance')

# Customizing the plot
plt.title('Student Attendance During the Week')
plt.xlabel('Day')
plt.ylabel('Number of Students Present')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Example 10.11
import pandas as pd
import matplotlib.pyplot as plt
# Test marks of a student in different subjects
data = pd.DataFrame({
    'Subjects': ['Maths', 'Science', 'English', 'History', 'Computer'],
    'Scores': [85, 90, 78, 88, 92]
})
# Bar chart with custom color and bar edge style
plt.bar(data['Subjects'], data['Scores'],
        color='skyblue', edgecolor='black')
# Customizing the plot
plt.title('Test Scores in Different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid only on Y-axis
plt.tight_layout()
plt.show()
```

```python
#Example 10.12
import pandas as pd
import matplotlib.pyplot as plt
# Creating sample data
data = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr','May'],
    'Enrollments': [50, 60, 55, 70, 65],
    'Average Marks': [75, 78, 80, 77, 82]
})
# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# First plot - Line chart for enrollments
axes[0].plot(data['Month'], data['Enrollments'], marker='o', color='blue')
axes[0].set_title('Monthly Enrollments')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Number of Students')
axes[0].grid(True)
# Second plot - Bar chart for average scores
axes[1].bar(data['Month'], data['Average Marks'], color='green')
axes[1].set_title('Average Marks')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Marks')
axes[1].grid(True)
# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Example 10.13
import matplotlib.pyplot as plt
# Sample data
students = ['Aman','Bina','Chetan','Divya']
maths_marks = [85, 92, 78, 88]
science_marks = [80, 91, 70, 90]
# Create 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# First subplot: Bar chart for Maths Scores
axes[0].bar(students, maths_marks, color='skyblue')
axes[0].set_title('Maths Marks')
axes[0].set_xlabel('Students')
axes[0].set_ylabel('Marks')
axes[0].grid(True)
# Second subplot: Pie chart for Science Marks
axes[1].pie(science_marks, labels=students, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Science Score Distribution')
# Improve spacing between plots
plt.tight_layout()
plt.show()
```
