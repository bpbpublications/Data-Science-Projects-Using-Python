```python
#Example 11.1
from sklearn.linear_model import LogisticRegression
# Training data
X = [[1], [2], [3], [4], [5]]  # Study hours
y = [0, 0, 0, 1, 1]            # 0 = Fail, 1 = Pass
# Create and train the model
model = LogisticRegression()
model.fit(X, y)
# Predict for 6 study hours
pred_output = model.predict([[6]])[0]
# Map the numeric result to text
result = "Pass" if pred_output == 1 else "Fail"
print(f"Predicted result for 6 hours of study: {result}")
```

```python
# Example 11.2
from sklearn.linear_model import LinearRegression
import numpy as np
# Study hours (input) and marks (output)
X = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([50, 60, 65, 70, 80])
# Create and train the model
model = LinearRegression()
model.fit(X, y)
# Predict marks for 7 study hours
predicted_marks = model.predict([[7]])
print("Predicted marks for 7 study hours:", predicted_marks[0])
```

```python
# Example 11.3
from sklearn import svm
# Example dataset: [weight (gms), size (cm)]
X = [
    [150, 7],   # Apple
    [170, 7.5], # Apple
    [140, 6.8], # Apple
    [130, 6.5], # Orange
    [120, 6.2], # Orange
    [110, 6.0]  # Orange
]
# Labels: 0 = Apple, 1 = Orange
y = [0, 0, 0, 1, 1, 1]
# Create and train the SVM model
model = svm.SVC(kernel='linear')
model.fit(X, y)
# Predict for a new fruit
prediction = model.predict([[145, 6.9]])[0]
# Output in human-readable form
if prediction == 0:
    print("The fruit is likely an Apple.")
else:
    print("The fruit is likely an Orange.")
```

```python
# Example 11.4
from sklearn.svm import SVC
import numpy as np
# Step 1: Input feature (study hours)
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
# Step 2: Output labels (Pass or Fail)
y = ['Fail', 'Fail', 'Fail', 'Fail', 'Pass', 'Pass', 'Pass', 'Pass']
# Step 3: Create SVM model with linear kernel
model = SVC(kernel='linear')
# Step 4: Train the model
model.fit(X, y)
# Step 5: Predict result for a new student
prediction = model.predict([[5.0]])
print("Predicted result:", prediction[0])
```

```python
#Example 11.5
from sklearn import tree
# Features: [Weather, Temperature]
# Weather: Sunny=0, Humid=1, Rainy=2
X = [
    [0, 25],
    [0, 22],
    [1, 20],
    [2, 18],
    [2, 15],
    [1, 23],
    [0, 19],
    [2, 21]
]
# Labels: Yes=1, No=0
y = [1, 1, 1, 0, 0, 1, 0, 0]
# Create Decision Tree model
clf = tree.DecisionTreeClassifier(criterion='gini')
clf = clf.fit(X, y)
# Test: Weather = Sunny (0), Temp = 20°C
prediction = clf.predict([[0, 20]])
if prediction[0] == 1:
    print("Decision: Play Outside!")
else:
    print("Decision: Stay Inside!")
```

```python
# Example 11.6
from sklearn.neighbors import KNeighborsClassifier
# Features (weight(gms), size(cm))
X = [[150, 7.0], [170, 7.5], [140, 6.8], [180, 7.8]]
# Labels: 0 = Apple, 1 = Orange
y = [0, 1, 0, 1]
# Create KNN model with K=3
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
# Predict fruit type for weight=145g, size=6.9cm
prediction = model.predict([[145, 6.9]])
if prediction[0] == 0:
    print("It's an Apple!")
else:
    print("It's an Orange!")
```

```python
# Example 11.7
# Naive Bayes: Predict if fruit is Apple or Orange based on Color
#  Priors
Prob_Apple = 1/2   # 1 Apple out of 2 fruits
Prob_Orange = 1/2  # 1 Orange out of 2 fruits
print("P(Apple) =", Prob_Apple)
print("P(Orange) =", Prob_Orange)
# Likelihoods
Prob_Red_given_Apple = 1  # Apple is red
Prob_Red_given_Orange = 0 # Orange is not red
print("\nP(Red | Apple) =", Prob_Red_given_Apple)
print("P(Red | Orange) =", Prob_Red_given_Orange)
# Posterior (Prior * Likelihood)
Prob_Apple_given_Red = Prob_Apple * Prob_Red_given_Apple
Prob_Orange_given_Red = Prob_Orange * Prob_Red_given_Orange
print("\nP(Apple | Red) =", Prob_Apple_given_Red)
print("P(Orange | Red) =", Prob_Orange_given_Red)
# Classify
if Prob_Apple_given_Red > Prob_Orange_given_Red:
    print("\nPredicted Fruit: Apple")
else:
    print("\nPredicted Fruit: Orange")
```

```python
# Example 11.8
# Import necessary libraries
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
# Actual labels
y_actual = ['Spam', 'Spam', 'Spam', 'Not Spam', 'Not Spam', 'Spam', 'Not Spam', 'Not Spam', 'Spam', 'Not Spam']
# Predicted labels by the model
y_predicted = ['Spam', 'Not Spam', 'Not Spam', 'Not Spam', 'Not Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam']
# Confusion Matrix
conf_mat = confusion_matrix(y_actual, y_predicted, labels=['Spam', 'Not Spam'])
print("Confusion Matrix:")
print(conf_mat)
# Accuracy
accuracy = accuracy_score(y_actual, y_predicted)
print("\nAccuracy:", accuracy)
# Precision
precision = precision_score(y_actual, y_predicted, pos_label='Spam')
print("Precision:", precision)
# Recall
recall = recall_score(y_actual, y_predicted, pos_label='Spam')
print("Recall:", recall)
# Step 7: F1-Score
f1 = f1_score(y_actual, y_predicted, pos_label='Spam')
print("F1-Score:", f1)
```

```python
# Example 11.9
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
# Actual labels (ground truth)
y_actual = ['Diabetes']*10 + ['Not Diabetes']*10   # 10 with diabetes, 10 without
# Predicted labels by the test (based on TP=8, FN=2, FP=2, TN=8)
y_predicted = ['Diabetes']*8 + ['Not Diabetes']*2 + ['Diabetes']*2 + ['Not Diabetes']*8
# Confusion Matrix
conf_matrix = confusion_matrix(y_actual, y_predicted, labels=['Diabetes', 'Not Diabetes'])
print("Confusion Matrix:")
print(conf_matrix)
# Performance Metrics
print("\nPerformance Metrics are:")
print("Accuracy :", accuracy_score(y_actual, y_predicted))
print("Precision:", precision_score(y_actual, y_predicted, pos_label='Diabetes'))
print("Recall   :", recall_score(y_actual, y_predicted, pos_label='Diabetes'))
print("F1-Score :", f1_score(y_actual, y_predicted, pos_label='Diabetes'))
```
