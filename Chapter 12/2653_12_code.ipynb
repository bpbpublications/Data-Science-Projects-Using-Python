```python
#Example 12.1: Clustering students by height & weight
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Step 1: Create data [Height(cm), Weight(kg)]
persons = np.array([
    [160, 55],  # Person A
    [165, 60],  # Person B
    [170, 65],  # Person C
    [180, 80],  # Person D
])

# Step 2: Create KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0)

# Step 3: Fit the model
kmeans.fit(persons)

# Step 4: Get cluster labels
labels = kmeans.labels_

# Step 5: Get centroids
centers = kmeans.cluster_centers_

# Step 6: Plot results
plt.scatter(persons[:, 0], persons[:, 1], c=labels, cmap='rainbow', s=100)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')  # plot centroids
plt.title("Persons grouped by Height & Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.show()
```

```python
# Example 12.2
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
# Step 2: Create a small dataset of (Math, Science) scores
scores = np.array([
    [83, 91], [84, 85], [93, 84],   # High performers
    [55, 45], [53, 45], [42, 51]    # Low performers
])
# Step 3: Apply K-Means clustering for 2 groups
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(scores)
# Step 4: Get the cluster labels
labels = kmeans.labels_
# Step 5: Get cluster centers
centers = kmeans.cluster_centers_
# Step 6: Plot the scores
plt.scatter(scores[:, 0], scores[:, 1], c=labels, cmap='coolwarm', s=100)
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='X', s=200, label='Centroids')
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("Clustering Student Scores")
plt.legend()
plt.show()
```

```python
# Example 12.3
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Step 1: Load dataset from CSV
df = pd.read_csv("F:\\iris.csv")
# Step 2: Select only the features (not the species column)
X = df.drop("species", axis=1)
# Step 3: Apply K-Means clustering (let’s assume 3 clusters since Iris has 3 species)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
# Step 4: Plot clusters (use first two features just for visualization)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=df['cluster'], cmap='viridis')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset clustered using K-Means")
plt.show()
```

```python
#Example 12.4
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Step 1: Small dataset with 2 subjects
data = {
    "Math": [80, 85, 78, 90, 88],
    "Science": [75, 82, 79, 92, 85]
}
df = pd.DataFrame(data)
print("Original Data:\n", df)
# Step 2: Apply PCA to reduce 2 features into 1 feature
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(df)
# Step 3: Create a new DataFrame
reduced_df = pd.DataFrame(X_reduced, columns=["PC1"])
print("\nReduced Data:\n", reduced_df)
# Step 4: Plot original vs reduced
plt.scatter(df["Math"], df["Science"], color="blue", label="Original Data")
plt.scatter(reduced_df.index, reduced_df["PC1"], color="red", label="Reduced (PC1)")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.title("PCA Dimensionality Reduction")
plt.show()
```

```python
#Example 12.5
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Step 1: A Small dataset with Height (cm) and Weight (kg)
data = {
    "Height": [150, 160, 170, 180, 190],
    "Weight": [50, 60, 65, 80, 90]
}
df = pd.DataFrame(data)
print("Original Data:\n", df)
# Step 2: Apply PCA to reduce 2 features into 1 feature
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(df)
# Step 3: Store reduced feature
reduced_df = pd.DataFrame(X_reduced, columns=["PC1"])
print("\nReduced Data:\n", reduced_df)

# Step 4: Plot
plt.scatter(df["Height"], df["Weight"], color="blue", label="Original Data")
plt.scatter(reduced_df.index, reduced_df["PC1"], color="red", label="Reduced (PC1)")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.title("Height & Weight Features converted into one Feature: PC1")
plt.show()
```

```python
# Example 12.6
import random

# Environment setup
states = [0, 1, 2, 3, 4]   # 0 = Start, 4 = Goal
goal_state = 4

# Agent starts at position 0
position = 0
rewards = []

print("Agent starts at position:", position)

# Run for 10 steps
for step in range(10):
    # Action: Move Left (-1) or Right (+1)
    action = random.choice([-1, 1])
    revised_position = position + action
    
    # Keep agent within bounds
    if revised_position < 0:
        revised_position = 0
    if revised_position > 4:
        revised_position = 4
    
    # Reward system
    if revised_position == goal_state:
        reward = 10
        print(f"Step {step+1}: Moved to {revised_position} Goal reached! Reward = {reward}")
    else:
        reward = -1
        print(f"Step {step+1}: Moved to {revised_position} Reward = {reward}")
    
    rewards.append(reward)
    position = revised_position  # Update agent's position
    
print("\nAll Rewards:", rewards)
print("Total Score:", sum(rewards))
```
